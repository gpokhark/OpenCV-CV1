```python
import os
```

```python
images = []

dirName = "scene"

imagefiles = [DATA_PATH + "images/{}/".format(dirName) + f for f in os.listdir(DATA_PATH+"images/"+dirName) if f.endswith(".jpg")]
imagefiles.sort()

destination = "{}_result.png".format(dirName)
plt.figure(figsize=[20,15])
i=1
for filename in imagefiles:
    img = cv2.imread(filename)
    images.append(img)

###
### YOUR CODE HERE
###
sticher = cv2.Stitcher_create(cv2.Stitcher_PANORAMA)
status, result = sticher.stitch(images)
```
