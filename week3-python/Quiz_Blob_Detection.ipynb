```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams['figure.figsize'] = [12.0, 12.0]
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
im = cv2.imread("data/images/5_new.png", cv2.IMREAD_GRAYSCALE)

detector = cv2.SimpleBlobDetector_create()

keypoints = detector.detect(im)

im = cv2.cvtColor(im, cv2.COLOR_GRAY2BGR)

for k in keypoints:
    x, y = k.pt
    x = int(round(x))
    y = int(round(y))
    cv2.circle(im, (x, y), 5, (255, 255, 255), -1)
    diameter = k.size
    radius = int(round(diameter / 2))
    cv2.circle(im, (x, y), radius, (0, 0, 255), 2)
plt.imshow(im[:, :, ::-1])
```

```python
params = cv2.SimpleBlobDetector_Params()
```

```python
params.minThreshold = 10
params.maxThreshold = 200

params.filterByArea = False
params.minArea = 1500

params.filterByCircularity = False
params.minCircularity = 0.1

params.filterByConvexity = False
params.minConvexity = 0.87

params.filterByInertia = True
params.minInertiaRatio = 0.01

detector = cv2.SimpleBlobDetector_create(params)

im = cv2.imread("data/images/6_new.png", cv2.IMREAD_GRAYSCALE)
keypoints = detector.detect(im)

im = cv2.cvtColor(im, cv2.COLOR_GRAY2BGR)

for k in keypoints:
    x, y = k.pt
    x = int(round(x))
    y = int(round(y))
    cv2.circle(im, (x, y), 5, (0, 0, 0), -1)
    diameter = k.size
    radius = int(round(diameter / 2))
    cv2.circle(im, (x, y), radius, (0, 0, 255), 2)
plt.imshow(im[:, :, ::-1])
```
